```python
import pandas as pd
import itertools
import warnings
warnings.filterwarnings('ignore')

df = pd.read_excel('dss_rules.xlsx')
df
```

```python
mappings = {
    'Last Avg Tensiometer': {
        'High': [500, 800],
        'Medium': [350],
        'Low': [0, 200]
    },
    'Predicted Avg Tensiometer': {
        'High': [500, 800],
        'Medium': [350],
        'Low': [0, 200]
    },
    'Predicted Rain Amount': {
        'High': [20, 50],
        'Medium': [10],
        'Low': [0, 5],
        'NaN': [2.5, 10, 35]
    },
    'Predicted Max Temperature': {
        'High': [35, 45],
        'Medium': [25],
        'Low': [0, 15],
        'NaN': [7.5, 25, 40]
    },
    'Decision': {
        'Not Recommended': [0],
        'Half Turn': [325],
        'Single Turn': [650],
        'Double Turn': [1300]
    }
}

expanded_data = []

for _, row in df.iterrows():
    last_avg_cat = row['Last Avg Tensiometer']
    predicted_avg_cat = row['Predicted Avg Tensiometer']
    predicted_rain_cat = row['Predicted Rain Amount'] if pd.notna(row['Predicted Rain Amount']) else 'NaN'
    predicted_max_temp_cat = row['Predicted Max Temperature'] if pd.notna(row['Predicted Max Temperature']) else 'NaN'
    decision_cat = row['Decision']

    predicted_rain_cat = str(predicted_rain_cat)
    predicted_max_temp_cat = str(predicted_max_temp_cat)

    last_avg_values = mappings['Last Avg Tensiometer'][last_avg_cat]
    predicted_avg_values = mappings['Predicted Avg Tensiometer'][predicted_avg_cat]
    predicted_rain_values = mappings['Predicted Rain Amount'][predicted_rain_cat]
    predicted_max_values = mappings['Predicted Max Temperature'][predicted_max_temp_cat]
    decision_values = mappings['Decision'][decision_cat]

    for last_val, pred_val, rain_val, temp_val, dec_val in itertools.product(
        last_avg_values, predicted_avg_values, predicted_rain_values, predicted_max_values, decision_values
    ):
        expanded_data.append([last_val, pred_val, rain_val, temp_val, dec_val])

expanded_df = pd.DataFrame(expanded_data, columns=[
    'Last Avg Tensiometer', 'Predicted Avg Tensiometer', 'Predicted Rain Amount', 'Predicted Max Temperature', 'Decision'
])

expanded_df.to_excel('train_data.xlsx', index=False)
print(f"Total Observations: {len(expanded_df)}")
expanded_df.head()
```
